## Exploratory Data Analysis (EDA)

```python
import  pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('power.csv')
```

```python
#hist for all colums 
df.hist(bins=30)
```

```python
#for clearness  
#in first column tem it is a bell curve
df[df.columns[0:4]].hist(bins=(20),figsize=(15,15))
plt.show()
```

```python
df[df.columns[1:8]].hist(bins=(40),figsize=(15,15))
plt.show()
```

```python
df[df.columns[1:9]].hist(bins=(40),figsize=(15,15))
plt.show()
```

```python
df[df.columns[9:18]].hist(bins=(40),figsize=(15,15))
plt.show()
```

```python
df[df.columns[18:20]].hist(bins=(10),figsize=(10,15))
plt.show()
```

```python
#Bivariate Analysis(2nd method)
#Scatter (which is better) with target feature

plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column],df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power(KW)')
    plt.xlabel(column)
    plt.ylabel('Generated Power(KW)')
plt.tight_layout()
plt.show()
```

```python
#corelation 
df.corr()
```

```python
#corelation with sns 
plt.figure(figsize=(15,10))
sns.heatmap(df.corr())
```

```python
#corelation with sns( to change colour)
sns.heatmap(df.corr(), cmap='coolwarm')
```

```python
 #to add numbers to it and fmt for geeting only two decimals
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt='.2f')
```

```python
#outlier
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[column])
plt.show()
```
